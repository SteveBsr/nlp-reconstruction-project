```python
from sentence_transformers import SentenceTransformer, util
import numpy as np
import pandas as pd

# Από το παραδοτέο 1B
text1 = [
    "Today is our dragon boat festival, in our Chinese culture, to celebrate it with all safe and great in our lives",
    "Thank your message to show our words to the doctor, as his next contract checking, to all of us",
    "I got this message to see the approved message",
    "In fact, I have received the message from the professor, to show me, this, a couple of days ago",
    "I am very appreciated the full support of the professor, for our Springer proceedings publication."
]

recon1_text1 = [
    "Today is our dragon boat festival in our Chinese culture to celebrate it with all safe and great in our lives.",
    "Thanks for your message to share our words with the doctor during his next contract review for all of us.",
    "I received this message to see the approved message.",
    "Actually, I got the message from the professor showing me this a few days ago.",
    "I am very grateful for the full support of the professor for the publication of our Springer proceedings."
]

recon2_text1 = [
    "The text discusses the celebration of the Dragon Boat Festival in Chinese culture and expresses gratitude for support received regarding a publication."
    "It also mentions communication with a professor and receiving messages about a contract and publication approval."
]

recon3_text1 = [
    "Today is our dragon boat festival, which we celebrate in our Chinese culture, with the goal of being safe and having a great time.",
    "Thank you for your message to show our words to the doctor as he checks the next contract for all of us.",
    "I got this message to see the approved one.",
    "In fact, I got the message from the professor a couple of days ago.",
    "I appreciate the full support of the professor for our Springer proceedings publication."
]

text2 = [
    "During our final discuss, I told him about the new submission — the one we were waiting since last autumn, but the updates was confusing as it not included the full feedback from reviewer or maybe editor.",
    "Anyway, I believe the team, although bit delay and less communication at recent days, they really tried best for paper and cooperation.",
    "We should be grateful, I mean all of us, for the acceptance and efforts until the Springer link came finally last week, I think.",
    "Also, kindly remind me please, if the doctor still plan for the acknowledgments section edit before he sending again.",
    "Because I didn’t see that part final yet, or maybe I missed, I apologize if so.",
    "Overall, let us make sure all are safe and celebrate the outcome with strong coffee and futuretargets."
]

recon1_text2 = [
    "During our final discussion, I told him about the new submission — the one we were waiting for since last autumn , but the updates were confusing as it did not include the full feedback from reviewer or maybe editor?",
    "Anyway, I think the team, although a bit delay and less communication at recent days, really tried for paper and cooperation best.",
    "We should be grateful, I mean all of us, for the acceptance and efforts until the Springer link finally came last week, I think.",
    "Also, kindly remind me please, if the doctor still plan for the acknowledgments section edit before he sends again.",
    "Because I didn’t see that part final yet, or maybe I missed it, I apologize if so.",
    "Let us make sure that all are safe and celebrate the outcome with strong coffee and future targets."
]

recon2_text2 = [
    "During our final discuss, I told him about the new link.",
    "But the updates was confusing as it not included the full feedback.",
    "I didn't see that part final yet, or maybe I missed, I apologize.",
    "Overall, let us make sure all are safe and celebrate the outcome with strong coffee."
]

recon3_text2 = [
    "During our final discuss, I told him about the new submission — the one we were waiting since last autumn, but the updates was confusing as it not included the full feedback from reviewer or maybe editor?",
    "Anyway, I believe the team, although bit delay and less communication at recent days, they really tried best for paper and cooperation.",
    "We should be grateful, I think all of us, for the acceptance and efforts until the Springer link came finally last week, I think, I think. We should be grateful, I think all of us, for the acceptance and efforts until the Springer link came finally last week, I think.",
    "Also, kindly remind me please, if the doctor still plan for the acknowledgments section edit before he sending again, kindly remind me please, if the doctor still plan for the acknowledgments section edit before he sending again.",
    "Because I didn’t see that part final yet, or maybe I missed, I apologize if so, I apologize if so. Because I didn’t see that part final yet, or maybe I missed, I apologize if so.",
    "Let us make sure all are safe and celebrate the outcome with strong coffee and future targets. Overall, let us make sure all are safe and celebrate the outcome with strong coffee and future targets. Let us make sure all are safe and celebrate the outcome with strong coffee and future targets."
]

# Μοντέλο similarity
similarity_model = SentenceTransformer('all-MiniLM-L6-v2')

sents1 = (text1)
sents2 = (text2)

# Συνάρτηση υπολογισμού similarity
def compute_similarity(original_sents, paraphrased_sents, model):
    similarities = []
    for orig, para in zip(original_sents, paraphrased_sents):
        emb1 = model.encode(orig, convert_to_tensor=True)
        emb2 = model.encode(para, convert_to_tensor=True)
        sim = util.pytorch_cos_sim(emb1, emb2).item()
        similarities.append(sim)
    return similarities, np.mean(similarities)

text1_joined = " ".join(sents1)
text2_joined = " ".join(sents2)
recon2_text1_joined = " ".join(recon2_text1)
recon2_text2_joined = " ".join(recon2_text2)

# Υπολογισμοί
sim1a, mean1a = compute_similarity(sents1, recon1_text1, similarity_model)
sim1b, mean1b = compute_similarity(sents1, recon3_text1, similarity_model)
sim1c, mean1c = compute_similarity([text1_joined], [recon2_text1_joined], similarity_model)

sim2a, mean2a = compute_similarity(sents2, recon1_text2, similarity_model)
sim2b, mean2b = compute_similarity(sents2, recon3_text2, similarity_model)
sim2c, mean2c = compute_similarity([text2_joined], [recon2_text2_joined], similarity_model)

# Πίνακας αποτελεσμάτων
df = pd.DataFrame({
    "Similarity Κείμενο 1": [mean1a, mean1c, mean1b],
    "Similarity Κείμενο 2": [mean2a, mean2c, mean2b]
})

print(df.round(4))
```
